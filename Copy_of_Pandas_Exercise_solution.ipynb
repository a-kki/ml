```python
## Question 9: Concatenate two data frames using following conditions
 ##   Create two data frames using following two Dicts, Concatenate those two data frame and create a key for each data frame.

##GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
## japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
```

```python
import pandas as pd # Import pandas

GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
carsDf1 = pd.DataFrame.from_dict(GermanCars) # Create DataFrame from German cars dict
print("German Cars DataFrame:")
print(carsDf1)

japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
carsDf2 = pd.DataFrame.from_dict(japaneseCars) # Create DataFrame from Japanese cars dict
print("\nJapanese Cars DataFrame:")
print(carsDf2)

combinedDf = pd.concat([carsDf1, carsDf2], keys=["German", "Japanese"]) # Concatenate DataFrames with keys
print("\nCombined DataFrame:")
combinedDf # Display concatenated DataFrame
```

```python
#Question 10: Merge two data frames using following condition
 #   Create two data frames using following two Dicts, Merge two data frames, and append second data frame as a new column to first data frame.

#Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
#car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
```

```python
import pandas as pd # Import pandas

Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
carPriceDf = pd.DataFrame.from_dict(Car_Price) # Create DataFrame from car price dict
print("Car Price DataFrame:")
print(carPriceDf)

car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
carsHorsepowerDf = pd.DataFrame.from_dict(car_Horsepower) # Create DataFrame from car horsepower dict
print("\nCar Horsepower DataFrame:")
print(carsHorsepowerDf)

carsDf = pd.merge(carPriceDf, carsHorsepowerDf, on="Company") # Merge DataFrames on 'Company'
print("\nMerged DataFrame:")
carsDf # Display merged DataFrame
```
