[1,2,3,4,5] → These are your x-values.

[1,4,9,16,25] → These are your y-values (squares of x-values).

'r^-' → This is a format string that styles the plot:

'r' → Red color

'^' → Triangle-up markers for each point

'-' → Solid line connecting the points

```python
plt.grid(True)
plt.title('squares of the numbers')
plt.xlabel('The Numbers', fontsize=18, color='red')
plt.ylabel('The squares',fontsize=18, color='red')
plt.plot([1,2,3,4,5], [1,4,9,16,25], 'r^-', linewidth=3.0)
plt.show()
```

### Interactive Stock Price Plot for Microsoft (MSFT)

Let's fetch a new dataset, Microsoft (MSFT) stock data, and visualize its closing price over time using an interactive Plotly line chart. This plot will include a range slider to easily zoom into specific time periods.

```python
import yfinance as yf
import plotly.graph_objects as go
import pandas as pd

# Download stock data for Microsoft (MSFT)
ticker_new = "MSFT"
start_date_new = "2023-01-01"
end_date_new = "2024-01-01"
data_new = yf.download(ticker_new, start=start_date_new, end=end_date_new)

# Reset index to use 'Date' as a column
data_new = data_new.reset_index()

print(f"Downloaded {len(data_new)} rows of data for {ticker_new} from {start_date_new} to {end_date_new}.")
display(data_new.head())
```

### Microsoft (MSFT) Stock Close Price Line Graph (Matplotlib)

Here's a simple line graph showing the closing price of Microsoft stock over the selected period, using `matplotlib`.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(data_new['Date'], data_new['Close'], color='purple', linewidth=2)
plt.title(f'{ticker_new} Stock Close Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.grid(True)
plt.xticks(rotation=45) # Rotate date labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```
