## Python Built-in Functions and Lambda Examples

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# Lambda: Anonymous function
x = lambda a, b : a * b
print(x(5, 6))
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# sum(iterable, start=0): Sums items of an iterable.
numbers = [1, 2, 3, 4, 5]
total = sum(numbers)
print(f"Sum of numbers: {total}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# map(func, iterable): Applies a specific function to every item in an iterable.
numbers = [1, 2, 3, 4]
squared_numbers = list(map(lambda x: x*x, numbers))
print(f"Squared numbers: {squared_numbers}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# filter(func, iterable): Creates an iterator of elements for which the function returns True.
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Even numbers: {even_numbers}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# reduce(func, iterable): (From functools) Iteratively applies a function to pairs of elements to reduce them to a single value.
from functools import reduce

numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print(f"Product of numbers: {product}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# sorted(iterable, key=...): Returns a new sorted list; the key argument often uses a lambda for custom sorting logic.
words = ["banana", "apple", "cherry"]
sorted_words = sorted(words)
print(f"Sorted words: {sorted_words}")

sorted_by_length = sorted(words, key=len)
print(f"Sorted by length: {sorted_by_length}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# reversed(iterable): Returns a reverse iterator without making a copy of the original object.
numbers = [1, 2, 3, 4, 5]
reversed_numbers = list(reversed(numbers))
print(f"Reversed numbers: {reversed_numbers}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# zip(*iterables): Pairs up items from multiple iterables into tuples for parallel iteration.
list1 = [1, 2, 3]
list2 = ['a', 'b', 'c']
zipped_lists = list(zip(list1, list2))
print(f"Zipped lists: {zipped_lists}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# enumerate(iterable, start=0): Adds a counter to an iterable, returning it as an enumerate object—highly useful for "for" loops.
fruits = ["apple", "banana", "cherry"]
for index, fruit in enumerate(fruits):
  print(f"Index {index}: {fruit}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# any(iterable): Returns True if any element in the iterable is true.
list_true = [True, False, False]
list_false = [False, False, False]
print(f"Any true in list_true: {any(list_true)}")
print(f"Any true in list_false: {any(list_false)}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# all(iterable): Returns True if all elements in the iterable are true.
list_true = [True, True, True]
list_mixed = [True, False, True]
print(f"All true in list_true: {all(list_true)}")
print(f"All true in list_mixed: {all(list_mixed)}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# next(iterator, default): Retrieves the next item from an iterator; can take a fallback value to avoid errors.
my_list = [10, 20, 30]
my_iterator = iter(my_list)
print(f"First item: {next(my_iterator)}")
print(f"Second item: {next(my_iterator)}")
print(f"Third item: {next(my_iterator, 'No more items')}")
print(f"Fourth item (with default): {next(my_iterator, 'No more items')}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# iter(object): Returns an iterator object for the given collection.
my_string = "hello"
string_iterator = iter(my_string)
print(f"Next char: {next(string_iterator)}")
print(f"Next char: {next(string_iterator)}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# len(object): Returns the length (number of items) of an object.
my_list = [1, 2, 3, 4, 5]
my_string = "hello"
print(f"Length of list: {len(my_list)}")
print(f"Length of string: {len(my_string)}")
```

```python
# Name-Akshay Khanna and Enrollment:0827CI231012
# max(iterable): Returns the largest item in an iterable.
# min(iterable): Returns the smallest item in an iterable.
numbers = [10, 5, 20, 15]
print(f"Max number: {max(numbers)}")
print(f"Min number: {min(numbers)}")
```
